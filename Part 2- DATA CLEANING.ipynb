# 2.0 Data Cleaning

```python
#This function will allow to remove any html tags

def html_removal(x):
    a=str(x)
    soup = BeautifulSoup(a)
    return soup.get_text(strip=True)
```

## Applying above function to column "body"

```python
df_subset=df_subset.copy()
```

```python
df_subset['body']=df_subset.body.apply(lambda x: html_removal(x))
```
